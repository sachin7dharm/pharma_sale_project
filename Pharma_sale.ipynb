```python
#creating a list of Medicines

l=list(df.columns)
l1=l[2:10]
l1
```

```python
# yearly sale

ydf=df.groupby('Year')[l1].sum().reset_index()
for medicine in l1:
    plt.plot(ydf['Year'],ydf[medicine],label=medicine,marker='o')
plt.title('Medicine Sales by Yearly')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend(loc='center')
plt.grid()
plt.show()
```

```python
# Average sale by yearly

avg_ydf=df.groupby('Year')[l1].mean().reset_index()
avg_ydf
```

```python

for medicine in l1:
    plt.plot(avg_ydf['Year'],avg_ydf[medicine],label=medicine,marker='o')
plt.title('Medicine Average Sale by Yearly')
plt.xlabel('Year')
plt.ylabel('Average Sales')
plt.legend()
plt.grid()
plt.show()
```

```python
#yearly month wise medicine sale

ymdf=df.groupby(['Year','Month'])[l1].sum().reset_index()
ymdf
```

```python
# M01AB medicine sale 

color={2014:'Green',2015:'Red',2016:'Blue',2017:'Brown',2018:'Orange',2019:'Purple'}

sns.catplot(x='Month',y=ymdf['M01AB'],data=ymdf,hue='Year',palette=color,s=40,linewidth=0.8)
plt.grid()
plt.show() 
```

```python
# total hourly sale from 2014 to 2019

hdf=df.groupby('Hour')[l1].sum().reset_index()
for medicine in l1:
    plt.plot(hdf['Hour'],hdf[medicine],label=medicine,marker='o')
plt.title('Medicine Sales by Hourly')
plt.xlabel('Hour')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()
```

```python
# total monthly sale from 2014 to 2019

mdf=df.groupby('Month')[l1].sum().reset_index()
for medicine in l1:
    plt.plot(mdf['Month'],mdf[medicine],label=medicine,marker='o')
plt.title('Medicine Sales by Month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()
```
